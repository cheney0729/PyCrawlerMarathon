1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests
url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'
r=requests.get(url)
response=r.text
import json
data=json.loads(response)
#json.loads(response)
print(len(data)) #物件的長度
l=[]
for key in data[0].keys():
    l.append(key)
print(l)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import requests
url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'
r=requests.get(url)
response=r.text

import json
data=json.loads(response)

#for i in range(len(data)):
    #print("標題:",data[i].get('title'))

for i in range(len(data)):
    j=data[i]
    print("標題:", j['title'])
    print("貼文時間:" ,j['createdAt'])
    print("留言人數:" ,j['commentCount'])
    print("按讚人數:", j['likeCount'])
    print('-------------------')
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

import requests
url='https://www.dcard.tw/_api/forums/pet/posts?popular=true'
r=requests.get(url)
response=r.text

import json
data=json.loads(response)

comment=0
like=0
print("熱門文章")

for i in range(len(data)):
    comment+=data[i].get('commentCount') #總留言人數
    like+=data[i].get('likeCount') #總按讚人數
print("平均留言人數:",comment/len(data))
print("平均按讚人數",like/len(data))


print()


url2='https://www.dcard.tw/_api/forums/pet/posts?popular=false'
r2=requests.get(url)
response=r2.text

data2=json.loads(response)

comment=0
like=0
print("非熱門文章")

for i in range(len(data2)):
    comment+=data2[i].get('commentCount') #總留言人數
    like+=data2[i].get('likeCount') #總按讚人數
print("平均留言人數:",comment/len(data))
print("平均按讚人數",like/len(data))
```
